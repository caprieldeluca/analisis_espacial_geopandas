```python
barrios = geopandas.read_file('datos/barrios.geojson')
barrios.head()
```

```python
barrios = barrios.to_crs(epsg=5345)
barrios.geometry
```

```python
m = barrios.explore(style_kwds={'color':'black', 'opacity':0.5, 'fill':False})
escuelas.explore(m=m, color='blue')
```

```python
villa_chartas = barrios.loc[barrios['nom_largo']=='Villa Chartas']
villa_chartas
```

```python
escuelas_villa_chartas = escuelas.sjoin(villa_chartas, how="inner", predicate='intersects')
escuelas_villa_chartas.head()
```

```python
m = villa_chartas.explore(style_kwds={'color':'black', 'opacity':0.5, 'fill':False})
escuelas_villa_chartas.explore(m=m, color='blue')
```

```python
escuelas_villa_chartas= escuelas_villa_chartas[['n_', 'nombre_del', 'domicilio', 'geometry']]
escuelas_villa_chartas
```

```python
buffers = escuelas_villa_chartas.buffer(100)
buffers
```

```python
escuelas_villa_chartas['buffers'] = buffers
escuelas_villa_chartas
```

```python
escuelas_villa_chartas.set_geometry('buffers', inplace=True)
escuelas_villa_chartas.geometry
```

```python
m = villa_chartas.explore(style_kwds={'color':'black', 'opacity':0.5, 'fill':False})
escuelas_villa_chartas.explore(m=m, style_kwds={'color':'blue', 'opacity':0.5, 'fill':True})
```

```python
villa_chartas_geom = villa_chartas.iloc[0]['geometry']
villa_chartas_geom
```

```python
relates = escuelas_villa_chartas.relate(villa_chartas_geom)
relates
```

```python
escuelas_villa_chartas_clipped = escuelas_villa_chartas.overlay(villa_chartas, how='intersection')

m = villa_chartas.explore(style_kwds={'color':'black', 'opacity':0.5, 'fill':False})
escuelas_villa_chartas_clipped.explore(m=m, style_kwds={'color':'blue', 'opacity':0.5, 'fill':True})
```

```python
escuelas_villa_chartas_clipped.unary_union
```

```python
escuelas_villa_chartas_clipped.unary_union.area
```

```python
escuelas_villa_chartas_clipped.dissolve()
```

```python
escuelas_villa_chartas_clipped.dissolve().area
```
